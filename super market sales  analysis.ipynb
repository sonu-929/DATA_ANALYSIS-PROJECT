# Feature Engineering :


```python
df=df.drop(columns=["Invoice ID","Time"])
df.head(5)
```

```python
df["Total_sells_price"]=df["Unit price"]*df["Quantity"]+df["Tax 5%"]+df["gross income"]
```

```python
continous=["Unit price","Tax 5%","Sales","cogs","gross margin percentage","gross income","Rating","Total_sells_price"] # for continous variable apply describe() function .
Discrete_categorical = ["Branch","City","Customer type","Gender","Product line","Date","Payment"]                      # for categorical variable apply unique() function . 
Discrete_count=["Quantity"]                                                                                            # for count variable apply value_counts() function .
```

```python


# Sample DataFrame (replace this with your actual data)
# df = pd.read_csv('your_data.csv')

columns = ["Unit price", "Tax 5%", "Sales", "cogs", "gross margin percentage", 
           "gross income", "Rating", "Total_sells_price"]

# Set plotting style
sns.set(style="whitegrid")

# Histograms
df[columns].hist(bins=15, color='skyblue', edgecolor='black', figsize=(14, 10), layout=(2, 4))
plt.suptitle('Histograms of Continuous Variables')
plt.tight_layout()
plt.show()

# Box Plots
df[columns].plot(kind='box', subplots=True, layout=(2, 4), figsize=(14, 10), color='lightcoral', title='Box Plots of Continuous Variables')
plt.tight_layout()
plt.show()

# Pair Plot
sns.pairplot(df[columns])
plt.suptitle('Pairplot of Continuous Variables', y=1.02)
plt.show()

# Correlation Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df[columns].corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()

# Scatter Plot: Unit Price vs Sales
sns.scatterplot(data=df, x='Unit price', y='Sales', color='purple')
plt.title('Unit Price vs Sales')
plt.xlabel('Unit Price')
plt.ylabel('Sales')
plt.show()

# Violin Plots
fig, axes = plt.subplots(2, 4, figsize=(14, 10))
for ax, column in zip(axes.flatten(), columns):
    sns.violinplot(data=df, x=column, color='lightgreen', ax=ax)
    ax.set_title(f'Violin Plot of {column}')
plt.tight_layout()
plt.show()
```
